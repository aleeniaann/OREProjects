```python
# you have to predict the price of the house based in given variable.
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df= pd.read_csv("data.csv")
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3, 2, 1500, 5000, 1, 0, 1, 3, 1200, 300, 1990, 2005]])
```
